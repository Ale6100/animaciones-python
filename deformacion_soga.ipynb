## Teoría

Si $q\ne3$, la evolución de la deformación de una soga ideal con extremos fijos sometida a una perturbación inicial en el centro en todo punto estará dada por:
$$\Psi(x,t)=\sum_{q=1}^{\infty}\frac{6h}{\pi(9-q^2)}\left(Sen(\frac{2\pi}{3}q)+Sen(\frac{\pi}{3}q)\right)Sen(\frac{q\pi}{L}x)Cos\left(\sqrt{\frac{T}{\mu}} \frac{q\pi}{L}t\right)$$

Mientras que si $q=3$ será:

$$\Psi(x,t)=-\frac{h}{3}Sen\left(\frac{3\pi}{L}x\right)Cos\left(\sqrt{\frac{T}{\mu}} \frac{3\pi}{L}t\right)$$
El siguiente código nos calcula y grafica el valor de $\Psi(x,t)$ para 12 valores de tiempos $t$ fijos:

```python
import numpy as np
import matplotlib.pyplot as plt

h = 0.5 # Perturbación inicial con amplitud 0.5m
L = 1 # Longitud de la cuerda de 1m
mu = 1 # Densidad lineal de la cuerda de 1kg/m^3
T = 1 # Tensión de la cuerda de 1N

t = np.linspace(0, 0.55, 12) # Creo una lista de tiempos
x = np.linspace(0, L, 1000) # Creo nuestro eje x (el eje horizontal de la soga que nunca cambia) que estará definido desde 0 hasta L

M = 50 # Cantidad de modos
for i in range(len(t)): # Quiero un gráfico para cada valor de tiempo t_o[i]
    Psi = 0
    for q in range(M+1): # Defino Psi considerando que el término del modo 3 es distinto
        if q == 3:
            Psi = Psi + (-h/3)*np.sin(3*np.pi*x/L)*np.cos(((T/mu)**(1/2))*3*np.pi*t[i]/L)  
        else:
            Psi = Psi + (6*h/(np.pi*(9 - q**2)))*(np.sin(2*np.pi*q/3)+np.sin(np.pi*q/3))*np.sin(q*np.pi*x/L)*np.cos(((T/mu)**(1/2))*q*np.pi*t[i]/L)
    plt.plot(x, Psi, label = 't = ' + str(round(t[i], 2))) # Grafico Psi en función de x para un valor t[i]

plt.ylabel('$\Psi$ (m)', fontsize=16)
plt.xlabel('x (m)', fontsize=16)
plt.suptitle('1° gráfico', fontsize=14, y=1)
plt.legend(loc = (1.01, 0))
plt.grid()
```

## Animación

A continuación se presenta una animación de la misma soga explicada en la teoría, con la diferencia de que acá se considerarán muchas más fotos para lograr una animación fluída.

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import animation, rc

rc('animation', html='html5')

h = 0.5 # Perturbación inicial con amplitud 0.5m
L = 1 # Longitud de la cuerda de 1m
mu = 1 # Densidad lineal de la cuerda de 1kg/m^3
T = 1 # Tensión de la cuerda de 1N

t = np.linspace(0, 8, 480) # Creo una lista de tiempos
x = np.linspace(0, L, 1000) # Creo nuestro eje x que estará definido desde 0 hasta L

Animacion = []

M = 50 # Cantidad de modos
for i in range(len(t)): # Quiero un gráfico para cada valor de tiempo t_o[i]
    Psi = 0
    for q in range(M+1): # Defino Psi considerando que el término del modo 3 es distinto
        if q == 3:
            Psi = Psi + (-h/3)*np.sin(3*np.pi*x/L)*np.cos(((T/mu)**(1/2))*3*np.pi*t[i]/L)  
        else:
            Psi = Psi + (6*h/(np.pi*(9 - q**2)))*(np.sin(2*np.pi*q/3)+np.sin(np.pi*q/3))*np.sin(q*np.pi*x/L)*np.cos(((T/mu)**(1/2))*q*np.pi*t[i]/L)
    Animacion.append(Psi) # Agrego Psi a la lista Animación

#ANIMACIÓN:
fig = plt.figure()                      
ax = fig.gca()                       

def actualizar(i):
    ax.clear()                           
    y = Animacion[i]
    plt.plot(x, y, color = 'blue') # Grafico, considerando el parámetro i
    plt.title('t = ' + str(round(t[i], 2)) + 's')
    plt.xlim(x[0], x[-1])
    plt.ylim(-h*1.1, h*1.1)
    plt.ylabel('$\Psi$ (m)', fontsize=16)
    plt.xlabel('x (m)', fontsize=16)
    plt.grid()

f = len(t)/(t[-1]-t[0])
print(f'Animación a {f} fotogramas por segundo')
X = 1000/f
ani = animation.FuncAnimation(fig, actualizar, range(len(Animacion)), interval = X)
ani
```
